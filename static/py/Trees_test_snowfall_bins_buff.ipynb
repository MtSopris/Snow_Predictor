```python
df = pd.read_csv(os.path.join("../data/bulk_data.csv"))
# df=df.replace([np.inf, -np.inf], np.nan, inplace=True)
df=df.fillna(0)
df.head()
```

```python
bins = [-340, -60, -12, -6, -3, 0,  3, 6, 12, 24, 60, 341]
labels = ['melting 28ft-5ft',' melting 5ft-1ft','melting 12in-6in', 'melting 6in-3in',
         'melting 3in-0in', '0in-3in', '3in-6in', '6in-12in', '12in-24in', '2ft-5ft','5ft-28ft']
df['binned'] = pd.cut(df['Change In Snow Depth (in)'], bins=bins, labels=labels)
df
    
```
